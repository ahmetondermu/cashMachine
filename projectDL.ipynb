```python
lastDataSet = newDataSet[newDataSet.year != 1970]
```

```python
lastDataSet.groupby("year").mean()["price"]
```

```python
lastDataSet = lastDataSet.drop("transmission" , axis = 1)
```

```python
lastDataSet
```

```python
y = lastDataSet["price"].values
x = lastDataSet.drop("price" , axis = 1).values
```

```python
print(x,"\n")
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.3 , random_state = 10)
```

```python
len(x_train)
```

```python
len(y_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(12 , activation = "relu"))
model.add(Dense(12 , activation = "relu"))
model.add(Dense(12 , activation = "relu"))
model.add(Dense(12 , activation = "relu"))

model.add(Dense(1))

model.compile(optimizer = "adam" , loss = "mse")
```

```python
model.fit(x = x_train , y = y_train , validation_data=(x_test , y_test) , batch_size=250 ,epochs = 300)
```

```python
lostData = pd.DataFrame(model.history.history)
```

```python
lostData.head()
```

```python
lostData.plot()
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
predictArray = model.predict(x_test)
```

```python
predictArray
```

```python
mean_absolute_error(y_test , predictArray)
```

```python
plt.scatter(y_test , predictArray)
plt.plot(y_test , y_test , "r-*")
```
